# EXPERIMENT 1 (Figure 4)

Reproducing Figure 4: relative accuracy and axiom satisfaction errors when training each architecture (MLP, CNN, WEC) on Plurality, Borda, and Copeland with IC sampling, up to 77 voters and 7 alternatives.

Common hyperparameters (paper §5.1):
* 77 voters, 7 alternatives, IC sampling
* 15,000 gradient steps, batch size 200, AdamW
* Cosine annealing with warm restarts (T_0 = 100)

Architecture specifics:
* MLP: 4 hidden layers à 128 neurons
* CNN: kernel1 = (5,1), kernel2 = (1,5), 64 channels, 3 linear layers à 128
* WEC: pretrained Word2Vec (corpus 100k, dim 200, window 7, skip-gram) + averaging + 3 linear layers à 128

```python
import exp1
import plot_and_visual

file_locations_mlp = {}
for rule_name in ['Plurality', 'Borda', 'Copeland']:
    print(f'Next: MLP with IC and rule {rule_name}')
    location = exp1.experiment1(
        architecture = 'MLP',
        rule_names = [rule_name],
        max_num_voters = 77,
        max_num_alternatives = 7,
        election_sampling = {'probmodel':'IC'},
        num_gradient_steps = int(15000),
        eval_dataset_size = int(500),
        sample_size_applicable = 100,
        sample_size_maximal = int(1e6),
        batch_size = 200,
        learning_scheduler = 100,
        comparison_rules = ['Plurality', 'Borda', 'Copeland'],
    )
    file_locations_mlp[rule_name] = location

plot_and_visual.plot_exp1(
    file_plurality = file_locations_mlp['Plurality'],
    file_borda = file_locations_mlp['Borda'],
    file_copeland = file_locations_mlp['Copeland'],
)
```
